```
from transformers import pipeline
import torch

# Suppose you have a pre-trained model in the folder:
model_path = "./Fine-tune/results/GPT 2/breathe-0/"  # adjust path as built/fine-tuned

generator = pipeline(
    "text-generation",
    model=model_path,
    tokenizer=model_path,
    device_map="auto"  # uses GPU if available
)

prompt = "Your input text here"
output = generator(prompt, max_new_tokens=30, repetition_penalty=1.3)
print(output[0]['generated_text'])
```
